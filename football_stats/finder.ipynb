```python
from mplsoccer import Sbopen
```

```python
parser = Sbopen()
```

```python
#finding tournament
parser.competition()
```

```python
#finding matches in tournament
df_matches = parser.match(competition_id=7, season_id=235)
df_matches
```

```python
#making full competition name
comp_name = f'{df_matches.competition_name.iloc[0]}, {df_matches.season_name.iloc[0]}'
comp_name
```

```python
#finding matches of the required team
cur_team = 'Paris Saint-Germain'
cur_team_m = df_matches[(df_matches['home_team_name'] == cur_team) | (df_matches['away_team_name'] == cur_team)]
cur_team_m
#All Ligue 1 matches 2022/2023 are matches involving PSG. This can be viewed by the sizes of the df_matches and cur_team_m tables.
```

```python
#finding the player whose statistics we want to look for
df_lineup = parser.lineup(3837876)
df_lineup[df_lineup.team_name == cur_team]
```

```python
cur_team_m_arr_ids = list(cur_team_m.match_id)
```

```python
#Due to the fact that Neymar often leaves with an injury, it is obvious that he does not participate in every match.
player_matches_ids = []
player_iden = 4320 #Neymar jr.
for i_id in cur_team_m_arr_ids:
    match_lineup = parser.lineup(int(i_id))
    if match_lineup['player_id'].isin([player_iden]).any():
        player_matches_ids.append(i_id)
```

```python
player_matches = cur_team_m[cur_team_m['match_id'].isin(player_matches_ids)]
```

```python
player_matches[['match_id', 'match_date', 'home_team_name', 'away_team_name', 'home_score', 'away_score']]
```

```python
df = parser.lineup(player_matches_ids[0])
player_nn = df.player_nickname[df.player_id == player_iden].iloc[0]
player_number = df.jersey_number[df.player_id == player_iden].iloc[0]
```
