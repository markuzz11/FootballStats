```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mpl

from mplsoccer import Sbopen

from variables import player_iden, player_nn, cur_team, comp_name, player_number, player_matches
```

```python
parser = Sbopen()
```

```python
match_id = 3837876
```

```python
event = pd.concat(parser.event(match_id))
```

```python
cols = ['id', 'period', 'minute', 'match_id', 'team_name', 'player_id', 'player_name', 'type_name', 'shot_statsbomb_xg', 'x', 'y', 'end_x', 'end_y']
shots_info = event[cols]
shots_info = shots_info[(shots_info['period'] <= 4) & (shots_info['type_name'] == 'Shot')]
shots_info
```

```python
player_info = shots_info[shots_info['player_id'] == player_iden]
print(f'{player_nn} xG = {player_info.shot_statsbomb_xg.sum()}')
```

```python
shots_info.groupby(['player_name', 'team_name'])['shot_statsbomb_xg'].sum().sort_values(ascending=False).head(5)
```

```python
team2 = list(set(shots_info.team_name.unique()) - {cur_team})[0]
```

```python
match_info = player_matches[player_matches.match_id == match_id]

home_team = cur_team if match_info.home_team_name.iloc[0] == cur_team else team2
home_team_goal = match_info.home_score.iloc[0]

away_team = cur_team if match_info.away_team_name.iloc[0] == cur_team else team2
away_team_goal = match_info.away_score.iloc[0]
```

```python
team1_time = np.array(shots_info[shots_info['team_name'] == cur_team]['minute'])
team1_xg = np.array(shots_info[shots_info['team_name'] == cur_team]['shot_statsbomb_xg'])

team2_time = np.array(shots_info[shots_info['team_name'] == team2]['minute'])
team2_xg = np.array(shots_info[shots_info['team_name'] == team2]['shot_statsbomb_xg'])

team1_time = np.insert(team1_time, 0, 0)
team1_xg = np.insert(team1_xg, 0, 0)
team2_time = np.insert(team2_time, 0, 0)
team2_xg = np.insert(team2_xg, 0, 0)
```

```python
team1_xg_progress = []
tmp = 0
for i in team1_xg:
  tmp += i
  team1_xg_progress.append(tmp)

team2_xg_progress = []
tmp = 0
for i in team2_xg:
  tmp += i
  team2_xg_progress.append(tmp)
```

```python
fig, ax = plt.subplots(figsize = (10,5))
fig.set_facecolor('#3d4849')
ax.patch.set_facecolor('#3d4849')

#set up our base layer
mpl.rcParams['xtick.color'] = 'white'
mpl.rcParams['ytick.color'] = 'white'

ax.grid(ls='dotted',lw=.5,color='lightgrey',axis='y',zorder=1)

spines = ['top','bottom','left','right']
for x in spines:
    if x in spines:
        ax.spines[x].set_visible(False)

plt.xlabel('Minute',color='white',fontsize=16)
plt.ylabel('xG',color='white',fontsize=16)
plt.ylim([0, max(round(team1_xg.sum()), round(team2_xg.sum()))+0.5])

ax.set_title(f'{home_team} : {away_team} \n ({home_team_goal} : {away_team_goal})',
             fontsize = 20, fontweight='bold', c='white')

#plot the step graphs
ax.step(x=team2_time,y=team2_xg_progress,color='#fd3607',label=team2,linewidth=3,where='post')
ax.step(x=team1_time,y=team1_xg_progress,color='yellow',label=cur_team,linewidth=3,where='post')
plt.legend()
```
